[# Notice how the subplot has multiple figures. Now let's add data to each axes.](https://)

````python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(2, 2, figsize=(10, 5))
X = np.linspace(0, 10, 100)  # 100   0  10
ax[0, 0].plot(X, X/2)
ax[0, 0].set_title("X vs X/2")
x_rand = np.random.rand(10)
y_rand = np.random.rand(10)
ax[0, 1].scatter(x_rand, y_rand)
ax[0, 1].set_title("Random Scatter")
favourite_food = {"Pizza": 4, "Burger": 7, "Pasta": 3, "Sushi": 5}
ax[1, 0].bar(favourite_food.keys(), favourite_food.values())
ax[1, 0].set_title("Favourite Food")
X_norm = np.random.randn(1000)  #   (Normal Distribution)
ax[1, 1].hist(X_norm, bins=30)
ax[1, 1].set_title("Histogram")
````

````python
import pandas as pd

car_sales = pd.read_csv("../data/car-sales.csv")

car_sales["Price"] = car_sales["Price"].str.replace('[\$,]', '', regex=True).str[:-2].astype(int)

-car_sales["Total Sales"] = car_sales["Price"].cumsum()

car_sales["Sale Date"] = pd.date_range(start=pd.Timestamp.today().normalize(),
                                       periods=len(car_sales),
                                       freq="D")

print(car_sales.head())

````

# Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

````python
import matplotlib.pyplot as plt

car_sales.plot(x="Sale Date", y="Total Sales", kind="line", figsize=(10, 6))

plt.title("Total Sales Over Time")
plt.xlabel("Sale Date")
plt.ylabel("Total Sales")

#  
plt.show()
````
